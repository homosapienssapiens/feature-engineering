```python
route = "/content/drive/MyDrive/Maestría en ciencia de datos/2do semestre/Ingeniería de características/Trabajo final/"
df = pd.read_csv(route + "data-1.csv")
```

Analizando el DataFrame podemos ver que las fechas no se encuentran en el formato fecha correcto. A continuación pondremos el formato a date time.

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df
```

```python
df['day'] = df['Fecha'].dt.day
df['month'] = df['Fecha'].dt.month
df['year'] = df['Fecha'].dt.year
df
```

```python
dfsplit1 = pd.DataFrame(columns = ['Date', 'x1', 'x2'])
x1 = float('NaN')
x2 = float('NaN')
for i in df['year'].unique().tolist(): 
  for j in df['month'].unique().tolist():
    for k in df['day'].unique().tolist():
      for l in df.index:
        if df.loc[l, 'year'] == i and df.loc[l, 'month'] == j and df.loc[l, 'day'] == k:
          fecha = df.loc[l, 'Fecha']  
          if df.loc[l, 'nombre'] == 'x1': 
            x1 = df.loc[l, 'data']
          else:
            x2 = df.loc[l, 'data']
      row = {'Date': fecha,
             'x1': x1,
             'x2': x2}
      dfsplit1 = dfsplit1.append(row, ignore_index = True)
      x1 = float('NaN')
      x2 = float('NaN')
dfsplit1
```

```python
dfsplit2 = pd.DataFrame(columns = ['Date', 'x1', 'x2'])
x1 = float('NaN')
x2 = float('NaN')
mydates = pd.date_range(min(df['Fecha']), max(df['Fecha'])).tolist()
for i in mydates:
  for j in df.index:
    if df.loc[j, 'Fecha'] == i:
      fecha = df.loc[j, 'Fecha']
      if df.loc[j, 'nombre'] == 'x1': 
        x1 = df.loc[j, 'data']
      else:
        x2 = df.loc[j, 'data']
  row = {'Date': fecha,
    'x1': x1,
    'x2': x2}
  dfsplit2 = dfsplit2.append(row, ignore_index = True)
  x1 = float('NaN')
  x2 = float('NaN')
dfsplit2
```
